```python
import pandas as pd
```

```python
dataset = pd.read_csv('train.csv')
```

```python
from pycaret.classification import *
s= setup(dataset,target = 'y', session_id = 8)
```

```python
compare_models(sort = 'F1')
```

Initially XGBoost was the top peforming model. Hence I went with the same.

```python
xgboost = create_model('xgboost')
```

```python
tuned = tune_model(xgboost,optimize = 'F1',choose_better = True,return_tuner = True,return_train_score=True)
```
